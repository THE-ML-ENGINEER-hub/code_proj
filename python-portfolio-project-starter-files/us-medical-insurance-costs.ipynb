Firstly, let us calculate the average age of the patience

```python
# Define a function called calculate average age that calculates the average age of the patience
def calculate_average_age(ages):
    total_ages = 0
    count_ages = 0
    for age in ages:
        total_ages += int(age)
        count_ages += 1
    average_age = round(total_ages / count_ages, 2)
    return average_age
average_age_patient = calculate_average_age(ages)
```

```python
print("The average age of patient is: " + str(average_age_patient) + " years old.")
```

Also, we would like to investigate the BMI of patient. 

We will find the average bmi of the patient. We define a functuion that will calculate the average bmi of the patient.

```python
# Define a funtion called calculate average bmis that calculates the average bmi of patience
def calculate_average_bmis(bmis):
    total_bmis = 0
    count_bmis = 0
    for bmi in bmis:
        total_bmis += float(bmi)
        count_bmis += 1
    average_bmis = round(total_bmis / count_bmis, 2)
    return average_bmis
average_bmis_patience = calculate_average_bmis(bmis)
    
```

```python
print("The Average BMIS of Patience is: " + str(average_bmis_patience) + " 25kg/m2.")
```

Let us investigate the bmis of the patients in depth. 
We will put the bmis of the patients on a scale.

```python
# Define a funtion called scale_bmis_patient
def scale_bmis_patient(bmis):
    bmis_range = []
    for bmi in bmis:
        if (float(bmi) < 18.5):
            bmis_range.append("underweigth")
        elif (float(bmi) == 18.5 and float(bmi) <= 24.9):
            bmis_range.append("normal weigth")
        elif (float(bmi) > 24.9 and float(bmi) <= 29.9):
            bmis_range.append("overweigth")
        else:
            bmis_range.append("obese")
    return bmis_range
range_bmis = scale_bmis_patient(bmis)
    
```

```python
print(range_bmis)
```

Let's investigate the insurance charges list that contains the medical insurance costs of patients.
We will find the average medical insurance cost of the patient

```python
# Define a function that calculates the average medical insurance cost of patient
def calculate_average_insurance_cost(insurance_charges):
    total_cost = 0
    count_cost = 0
    for cost in insurance_charges:
        total_cost += float(cost)
        count_cost += 1
    average_insurance_cost = round(total_cost / count_cost)
    return average_insurance_cost
average_insurance_cost_patient = calculate_average_insurance_cost(insurance_charges)
```

```python
print("The Average Medical Insurance Cost of Patient is: " + str(average_insurance_cost_patient) + " dollars.")
```

Let us also find the total number of males and females.
Will define a function that will calculste the fiigures

```python
# Define a funtion sex analyzer
def analyze_sex(sex):
    males = 0
    females = 0
    for gender in sex:
        if (gender == "male"):
            males += 1
        elif gender == "female":
            females += 1
    return f"The number of males is: {males}. \nThe number of females is: {females}"
    
gender_count = analyze_sex(sex)
```

let us find the average age of someone who has atleast one child

```python
# Define a function that calculate the average child of patient
def calculate_child_analyzer(num_children):
    total_children = 0
    count_children = 0
    for child in num_children:
        if int(child) > 0:
            total_children += int(child)
            count_children += 1
    average_child = round(total_children / count_children, 2)
    return average_child
average_children= calculate_child_analyzer(num_children)
```

```python
print(f"The average child of patient is: {average_children}")
```

Let us figure out what the average age is for someone who has at least one child in the dataset

```python
# Define a function that analyzes the average age of a patient with at least one child
def patient_average_age_per_child(ages, num_children):
    total_ages = 0
    count_ages = 0
    for age, child in zip(ages, num_children):
        if int(child) > 0:
            total_ages += int(age)
            count_ages += 1
    average_age = round(total_ages / count_ages, 2)
    return average_age
age_averages = patient_average_age_per_child(ages, num_children)
```

```python
print(f"The average age for someone with at least one child is: {age_averages} years old.")
```

```python
# Define a function that determines the unique regions the patients are located
def patients_unique_regions(regions):
    unique_regions = []
    for region in regions:
        if (region not in unique_regions):
            unique_regions.append(region)
    return unique_regions
patients_regions = patients_unique_regions(regions)
      
```

```python
print(patients_regions)
```

```python
# Define a function that calculates the different costs between smokers and non_smokers
def analyze_costs_different_smokers(smoker, insurance_charges):
    smokers_cost = 0
    non_smokers_cost = 0
    for smoke, insurance_charge in zip(smoker, insurance_charges):
        if (smoke == "yes"):
            smokers_cost += float(insurance_charge)
        elif (smoke == "no"):
            non_smokers_cost += float(insurance_charge)
    print("smokers cost is: " + str(round(smokers_cost, 2)))
    print("non smokers cost is: " + str(round(non_smokers_cost, 2)))
costs = analyze_costs_different_smokers(smoker, insurance_charges)
```

```python
# Create a dictionary 
def create_dictionary(ages, sex, bmis, num_children, smoker, insurance_charges):
    patients_dictionary = {}
    patients_dictionary["ages"] = [int(age) for age in ages]
    patients_dictionary["sex"] = [gender for gender in sex]
    patients_dictionary["bmis"] = [float(bmi) for bmi in bmis]
    patients_dictionary["num_children"] = [int(child) for child in num_children]
    patients_dictionary["smoker"] = [smoke for smoke in smoker]
    patients_dictionary["regions"] = [region for region in regions]
    patients_dictionary["insurance_charges"] = [float(charge) for charge in insurance_charges]
    return patients_dictionary 
patients_medical_records = create_dictionary(ages, sex, bmis, num_children, smoker, insurance_charges)
```
